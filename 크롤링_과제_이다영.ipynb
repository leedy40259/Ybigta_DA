```python
import time
import openpyxl
```

```python
chrome_driver = '/Users/dayounglee/Desktop/Ybigta/chromedriver'
driver = webdriver.Chrome(chrome_driver)
```

```python
from selenium.common.exceptions import NoSuchElementException
lst = []

for i in range(1, 15):
    url = f'http://www.yes24.com/Product/communityModules/GoodsReviewList/110791050?Sort=1&PageNumber={i}&Type=ALL&_=1665772671164'
    driver.get(url)
    driver.implicitly_wait(10)
    for j in range(2, 7):
        try:
            review = driver.find_element(By.XPATH, f'//*[@id="infoset_reviewContentList"]/div[{j}]/div[3]/div[2]').text
            review = review.replace('\n', ' ')
            lst.append(review)
        except NoSuchElementException:
            break
            
        
print(lst)
```

```python
wb = openpyxl.Workbook()
sheet = wb.active
```

```python
for i in range(len(lst)):
    sheet.append([lst[i]])
    
print(sheet)
```
